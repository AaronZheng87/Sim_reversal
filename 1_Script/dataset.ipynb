```python
import random
import pandas as pd
import numpy as np

def generate_random_sequence(min_val, max_val, total_sum, length):
    if length <= 0 or total_sum < min_val * length or total_sum > max_val * length:
        return None

    sequence = []
    remaining_sum = total_sum

    for i in range(length - 1):
        upper_bound = min(max_val, remaining_sum - min_val * (length - 1 - i))
        lower_bound = max(min_val, remaining_sum - max_val * (length - 1 - i))
        value = random.randint(lower_bound, upper_bound)
        sequence.append(value)
        remaining_sum -= value

    sequence.append(remaining_sum)  # Ensure the last value makes the sum correct

    return sequence

# Generate a random sequence with given constraints
min_value = 4
max_value = 14
total_sum = 24
sequence_length = 3

```

```python



random_sequence = generate_random_sequence(min_value, max_value, total_sum, sequence_length)
result=[]
for idx,value in enumerate(random_sequence,start=1):
    for n in range(1,value+1,1):
        if n ==1:
            result.append([f'e{idx}',n, 'reversed'])
        else:
            result.append([f'e{idx}',n, 'stable'])
data=pd.DataFrame(result,columns=['Episode','Trial_idx_epi', 'State'])
```

```python
initial_key = np.array([0,1])
random.shuffle(initial_key)

data['Correct_response'] = "None"
data.at[0, 'Correct_response'] = initial_key
for i in range(0, len(data["State"])):
    if data.iloc[i].State == 'reversed':
        initial_key = 1-initial_key
    if data.iloc[i].Stim == 'face':
        data.at[i, 'Correct_response'] = initial_key[0]
    else:
        data.at[i, 'Correct_response'] = initial_key[1]
data["Trial_id"] = range(1,len(data)+1)
```
